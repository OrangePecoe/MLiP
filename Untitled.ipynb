```python
import matplotlib.pyplot as plt
import csv

x=[]
y=[]

with open('data_a.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))


plt.plot(x,y, linewidth=0.5)
#ya = y[1]
#yb = y[10000-2]

ya = 102500
yb = 97500

axes = plt.gca()
axes.set_xlim([0,10000])
axes.set_ylim([yb,ya])


plt.xlabel('n')
plt.ylabel('Difference')

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
fig.savefig('fig_exca', dpi=100)

plt.show()
```
